## Joining into one dataframe to input to the model

```python
# Get last gameweek we have data for
gws_paths = os.listdir(get_data_path(SEASON, "gws"))
sorted_gws_paths = sorted(gws_paths, key=lambda x: int(x.split(".")[0].split("gw")[1]), reverse=True)
last_gw = int(sorted_gws_paths[0].split(".")[0].split("gw")[1])
last_gw
```

```python
# Merge columns into the fpl_gw_df
# key_passes, passes_into_penalty_area, crosses_into_penalty_area

for gw_num in range(1, last_gw + 1):
    fpl_gw_df = pd.read_csv(get_data_path(SEASON, f"gws/gw{gw_num}.csv")).drop(columns=["fixture", "value", ])
    fbref_gw_df = pd.read_csv(get_data_path(SEASON, f"fbref/gws_passing/gw{gw_num}.csv"))

    fpl_gw_df = fpl_gw_df.merge(fbref_gw_df[["player_id", "key_passes", "passes_into_penalty_area", "crosses_into_penalty_area"]],
                                on="player_id", how="left")

fpl_gw_df.fillna(0, inplace=True) # Null means player didnt play
fpl_gw_df.head()
```
