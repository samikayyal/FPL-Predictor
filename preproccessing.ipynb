```python
merged_stat_df = None

for file in os.listdir('stat_data'):
    if file.endswith(".csv"):
        stat_df = pd.read_csv(os.path.join('stat_data', file))
        if 'web_name' in stat_df.columns:
            stat_df.drop(columns=['web_name'], inplace=True)

        if merged_stat_df is None:
            merged_stat_df = stat_df

        if stat_df.duplicated(subset=['player_id', 'gw']).any():
            print(f"Duplicate entries found in {file}.")
            continue

        merged_stat_df = pd.merge(merged_stat_df, stat_df, on=['player_id', 'gw'], how='outer')

merged_stat_df.info()
```

## Joining into one dataframe to input to the model

```python
gw_df_list = []

for gw_num in range(1, LAST_PLAYED_GAMEWEEK + 1):
    gw_file_path = get_data_path(SEASON, "gws", f"gw{gw_num}.csv")
    gw_df = pd.read_csv(gw_file_path)
    gw_df.rename(columns={"gameweek": "gw"}, inplace=True)

    gw_stat_df = merged_stat_df[merged_stat_df.gw == gw_num]
    if gw_stat_df.duplicated(subset=['player_id', 'gw']).any():
        print(f"Warning: Duplicates found for (player_id, gw) in {gw_file_path}")
        continue

    gw_df = pd.merge(gw_df, gw_stat_df, on=["player_id", "gw"], how="left")
    gw_df_list.append(gw_df)

merged_gw_df = pd.concat(gw_df_list, ignore_index=True)
if merged_gw_df.duplicated(subset=['player_id', 'gw']).any():
    print("Warning: Duplicates found in the merged gameweek DataFrame.")
    print(merged_gw_df.duplicated(subset=['player_id', 'gw']).sum())
```

```python
# Add passing data into the merged gameweek DataFrame
# first merge the passing data to one df
passing_df = None
for file in os.listdir(get_data_path(SEASON, "fbref", "gws_passing")):
    if file.endswith(".csv"):
        gw_num = int(file.split(".")[0][2:])
        gw_passing_df = pd.read_csv(get_data_path(SEASON, "fbref", "gws_passing", file))
        gw_passing_df = gw_passing_df[['player_id', 'gw', 'key_passes',
                                       'total_pass_completion_percentage',]]

        if passing_df is None:
            passing_df = gw_passing_df
        else:
            passing_df = pd.concat([passing_df, gw_passing_df], ignore_index=True)

passing_df.fillna(0, inplace=True)
passing_df.info()
```

```python
merged_gw_df = pd.merge(merged_gw_df, passing_df, on=["player_id", "gw"], how="left")
if merged_gw_df.duplicated(subset=['player_id', 'gw']).any():
    print("Warning: Duplicates found in the merged gameweek DataFrame after passing merge.")
    print(merged_gw_df.duplicated(subset=['player_id', 'gw']).sum())
```
